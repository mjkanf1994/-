```python
import requests
import re
from bs4 import BeautifulSoup
url = 'https://www.yna.co.kr/view/AKR20200204081000017?section=it/it'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup)
```

```python
div = soup.find("div", {"class" : "story-news article"});div
```

```python
##string화 시키기
p_tag=div.find_all('p')
content=''
for i in p_tag:
    content+=i.text
content[:500]    
```

```python
content
```

```python
print(content.split(' '))
```

```python
from konlpy.tag import Komoran
komoran=Komoran()
komoran_tokens=komoran.morphs(content)
print(komoran_tokens)
```

```python
from konlpy.tag import Hannanum
hannanum= Hannanum( )
hannanum_tokens = hannanum.morphs(content) 
print(hannanum_tokens) 
```

```python
from konlpy.tag import Okt
okt= Okt()  
okt_tokens = okt.morphs(content) 
print (okt_tokens) 
```

```python
from konlpy.tag import Kkma 
kkma= Kkma() 
kkma_tokens = kkma.morphs(content)  
print (kkma_tokens) 
```

```python
#품사부착
```

```python
komoranTag = []
for token in komoran_tokens: 
    komoranTag += komoran.pos(token)
print (komoranTag) 
```

```python
hannanumTag = [] 
for token in hannanum_tokens: 
    hannanumTag += hannanum.pos(token)
print(hannanumTag)
```

```python
oktTag = [] 
for token in okt_tokens:
    oktTag += okt.pos(token)
print(oktTag)
```

```python
kkmaTag = []
for token in kkma_tokens:
    kkmaTag += kkma.pos(token)
print (kkmaTag)
```

```python
#최빈어 조회. 불용어를 선택하기 위한 단계
from collections import Counter
Counter (oktTag).most_common() 
```

```python
#불용어 처리
stopPos = ['Punctuation', 'Adverb','Conjunction', 'Josa', 'PreEomi', 'Eomi', 'Suffix','Punctuation', 'Foreign', 'Alpha', 'Number', 'Unknown']
stopWord = ['할', '수', '있게', '된다', '설명', '했다','서울','연합뉴스','홍','지인','기자','ljungberg@yna.co.kr', '저작권', '자', '무단', '전', '재', '재', '배포', '금지', '송', '고']
word = []
for tag in oktTag: 
    if tag[1] not in stopPos:
        if tag[0] not in stopWord:
            word.append(tag [0])
print (word) 
```

```python
from konlpy.tag import Mecab
m= Mecab()
m_tokens=m.morphs(content)
print(m_tokens)
```

```python
#품사부착
mecabTag = [] 
for token in m_tokens:
    mecabTag += m.pos(token)
print(mecabTag)
```

```python
#최빈어 조회. 불용어를 선택하기 위한 단계
from collections import Counter
Counter (mecabTag).most_common() 
```

```python
#불용어 처리
stopPos = ['SF', 'JKO','MAG', 'JX','SY','JKB','EC','SY','VV+EP','SSC','VA','SSO','SN','VV+EC']
stopWord = ['은','서울', '연합뉴스', '홍지인', '기자','의','무단', '전재', '재', '배포', '금지', '/', '/', ':', '송고','가','ljungberg', 'yna', 'co', 'kr', '저작권자', 'c', ',']
word = []
for tag in mecabTag: 
    if tag[1] not in stopPos:
        if tag[0] not in stopWord:
            word.append(tag [0])
print (word) 
```
